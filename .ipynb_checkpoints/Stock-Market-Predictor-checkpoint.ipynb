```python
# Import the libraries
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import math
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
plt.style.use('fivethirtyeight')
```

```python
# Get the stock quote
df = web.DataReader('AAPL', data_source = 'yahoo', start = '2012-01-01', end = '2019-12-27')

# Show the data
df
```

```python
df.shape
```

```python
# Visualize the closing price history
plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price USD ($)', fontsize = 18)
plt.show()
```

```python
# Create a new dataframe with only Close column
data = df.filter(['Close'])

# Training data length
train_data_len = math.ceil(len(data.values) * 0.8)
train_data_len
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range = (0,1))
scaled_data = scaler.fit_transform(data.values)

scaled_data
```

Create the training data set


```python
# Create the scaled training data set

train_data = scaled_data[0 : train_data_len, :]
x_train = []
y_train = []
# Choose 60 days because it gave the lease rmse 
for i in range(60, len(train_data)):
    # Independent variable set 'X',i.e, holds the stock price value for the past 60 days at position 'i-60'
    x_train.append(train_data[i-60:i, 0])
    # Dependent variable 'Y', i.e, holds the stock price at the 61st day at position 'i-60'
    y_train.append(train_data[i, 0])
```

```python
# Converting to numpy arrays for training the model
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the data so that LSTM model accepts it
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# Building the LSTM model
model = Sequential()

# First layer: return sequence is True since we are adding more layers to the model
model.add( LSTM(units = 50, return_sequences = True, input_shape = (x_train.shape[1], 1)))
# Next Layer: return sequence is False by default
model.add( LSTM( units = 50))
# Densely connected neural network layer with 25 neurons
model.add( Dense( units = 25))
model.add( Dense( units = 1))
```

```python
model.compile( optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train,y_train, batch_size = 1, epochs = 1)
```

Creating the testing data set

```python
test_data = scaled_data[train_data_len - 60 : , : ]
x_test = []
y_test = data.values[train_data_len : , : ]
for i in range(60, len(test_data)):
    x_test.append( test_data[i-60:i, 0])
x_test = np.array(x_test)
```

```python
x_test = np.reshape( x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
# Getting the models predicted price values
pred = model.predict(x_test)
pred = scaler.inverse_transform(pred) # Removing scaling
```

```python
rmse = np.sqrt( np.mean((pred - y_test) ** 2))
rmse
```

```python
# Plotting the data
train = data[:train_data_len]
valid = data[train_data_len:]
valid['Pred'] = pred

plt.figure( figsize = (16,8))
plt.title('Model')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price USD ($)', fontsize = 18)
plt.plot(train['Close'])
plt.plot(valid[['Close','Pred']])
plt.legend(['Train','Val','Pred'], loc = 'upper left')
plt.show()
```

```python
#Get the quote from '2012-01-01' to '2019-12-27' to predict the closing price for '2019-12-28'
new_df = df.filter(['Close']) 
days = new_df[-60:].values 
days_scaled  = scaler.transform(days) 
X_test = [] 
X_test.append(days_scaled) 
X_test = np.array(X_test) 
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

pred_price = model.predict(X_test) 
#Undo the scaling
pred_price = scaler.inverse_transform(pred_price) 
print("Predicted Price:",pred_price)
today = '2019-12-28'
todays_price = web.DataReader('AAPL', data_source='yahoo', start=today, end=today)
print("Actual Closing Price:", todays_price['Close'].values)
```
